```python
import optuna
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
# no categorical_feature
# best: 0.930834
#study_name="lightgbm_02"

# with categorical_feature
# old best: 0.929955 @ 417
# best 0.930021 @ 1926
#study_name="lightgbm_03"

# simple version no categorical_feature
study_name="lightgbm_04"

study = optuna.create_study(
    study_name=study_name,
    storage="sqlite:///optuna.db",
    load_if_exists=True,
    direction='maximize'
)
df = study.trials_dataframe()
df.sort_values('value', ascending=False, na_position='first', inplace=True)
print(len(df))
df
```
