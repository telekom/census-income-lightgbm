```python
import optuna
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
study_name="lightgbm_simple_01"

study = optuna.create_study(
    study_name=study_name,
    storage="sqlite:///optuna.db",
    load_if_exists=True,
    direction='maximize'
)
df = study.trials_dataframe()
df.sort_values('value', ascending=False, na_position='first', inplace=True)
print(len(df))
df
```

```python
!mkdir -p plot
```

```python
fig = optuna.visualization.plot_slice(study)
fig.write_html(f"plot/{study_name}_plot_slice.html")
```

```python
fig = optuna.visualization.plot_optimization_history(study)
fig.write_html(f"plot/{study_name}_plot_optimization_history.html")
```

```python
fig = optuna.visualization.plot_param_importances(study)
fig.write_html(f"plot/{study_name}_plot_param_importances.html")
```
